Bounding Box Approach

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

# Bounding box coordinates for Kassel
north = 51.369403
south = 51.260381
east = 9.570084
west = 9.351023

# Define tags to retrieve specific OSM features (optional)
tags = {"boundary":"postal_code"}

# Create a GeoDataFrame of OSM features within the specified bounding box
gdf = ox.features.features_from_bbox(north, south, east, west, tags)

# Plot the GeoDataFrame
gdf.plot()
plt.title('OSM Features in Kassel Bounding Box')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

Landkreis & Dokumenta Stadt Kassel Admin LVL 6 Tag

```python
place_name = "Kassel, Hesse, Germany"
tags = {"admin_level":"6" }

admin_6_method =ox.features_from_place(place_name,tags)

admin_6_method[48:]
```

Kassel Polygon stimmt: Row 49, Admin LVL 6 Tag

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

# Set up the plotting environment
fig, ax = plt.subplots(figsize=(10, 10))

# Extract the geometries for rows 49 
geometry_49 = admin_6_method.iloc[49]['geometry']

# Plot the geometries
gpd.GeoSeries([geometry_49]).plot(ax=ax, alpha=0.5)

# Show the plot
plt.show()
```

```python
data = admin_6_method.iloc[49]
data
```
