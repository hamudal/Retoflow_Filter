Landkreis & Dokumenta Stadt Kassel Data

```python
place_name = "Kassel"
tags = {"admin_level":"6" }

admin_6 =ox.features_from_place(place_name,tags)

admin_6
```

Kassel Polygon stimmt: Row 49, Admin LVL 6 Tag

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

# Set up the plotting environment
fig, ax = plt.subplots(figsize=(10, 10))

# Extract the geometries for rows 49 
geometry_49 = admin_6.iloc[49]['geometry']

# Plot the geometries
gpd.GeoSeries([geometry_49]).plot(ax=ax, alpha=0.5)

# Show the plot
plt.show()
```
