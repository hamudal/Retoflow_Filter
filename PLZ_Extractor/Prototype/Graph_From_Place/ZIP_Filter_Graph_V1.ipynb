```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

postal_code = "34131"
city = "Kassel"

# Get city boundary
city_boundary = ox.geocode_to_gdf({"city": city, "postal_code"postal_code:}, which_result=None).unary_union

# Extract postal code geometry
search_tags = {"postal_code": postal_code}
features = ox.features.features_from_polygon(city_boundary, search_tags)
postal_code_geometry = gpd.GeoDataFrame(geometry=features['geometry'], data=None, crs=None)

# Plot postal code geometry
postal_code_geometry.boundary.plot(color="black", linewidth=1.5)
plt.title("Postal Code Shape")
plt.gca().set_facecolor('white')
plt.show()
```

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

# Define the postal code and city
postal_code = "34131"
city = "Kassel, Germany"

# Fetch postal code boundary data directly within the city context
features = ox.geocode_to_gdf(postal_code, proximity=city)

# Plot the postal code boundary
gdf = gpd.GeoDataFrame(geometry=features['geometry'], data=None, crs=None)
gdf.boundary.plot(color="black", linewidth=1.5)
plt.title(f"Postal Code {postal_code} Boundary Shape")
plt.show()
```

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

# Get the polygon shape of Kassel
city_polygon = ox.geocode_to_gdf("Kassel, Germany")

# Define tags for OSM features
tags = {'building': True, 'landuse': True, 'highway': True}

# Get OSM features within the boundaries of the city polygon
features = ox.features.features_from_polygon(city_polygon['geometry'].iloc[0], tags)

# Create a GeoDataFrame from the features
features_gdf = gpd.GeoDataFrame(features)

# Plot the features
ax = features_gdf.plot(alpha=0.7, figsize=(10, 10), edgecolor='k')
city_polygon.boundary.plot(ax=ax, color='red', linewidth=2)
plt.title("OSM Features within Kassel Boundary")
plt.show()
```
