```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

def get_zip_features(place, tags):
    return ox.features_from_place(place, tags)

def filter_zip_code(gdf, postal_code):
    return gdf[gdf["postal_code"] == postal_code]

def plot_zip_code(zip_gdf):
    zip_gdf.boundary.plot(color="black", linewidth=2)
    plt.title(f'Postal Code {zip_gdf["postal_code"].iloc[0]} in Kassel')
    plt.gca().set_facecolor('white')
    plt.show()

tags_zip = {"boundary": "postal_code", "postal_code": "*" }

all_zips = get_zip_features("Kassel, Germany", tags_zip)
zip_34131 = filter_zip_code(all_zips, "34131")

if not zip_34131.empty:
    plot_zip_code(zip_34131)
else:
    print("No features found for the specified postal code.")
```

```python
import osmnx as ox
from geopy.geocoders import Nominatim
import matplotlib.pyplot as plt

def plot_postal_code(postal_code):
    # Erstellt einen String, der den Ort und die Postleitzahl repräsentiert
    place = f"{postal_code}, Germany"

    # Verwendet die geocode Funktion von osmnx, um die Koordinaten der Postleitzahl zu erhalten
    location = ox.geocode(place)

    # Verwendet die geolocator Funktion von geopy, um die Grenzen der Postleitzahl zu erhalten
    geolocator = Nominatim(user_agent="geoapiExercises")
    location = geolocator.geocode(place)
    gdf = ox.gdf_from_place(location.raw['osm_id'], which_result=2)

    # Zeichnet die Grenzen der Postleitzahl
    fig, ax = ox.plot_shape(gdf, figsize=(3,3))
    plt.show()

plot_postal_code("34131")
```
