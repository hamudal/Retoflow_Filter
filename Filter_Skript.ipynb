```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

def plot_postal_code_shape(city_name, postal_code):
    # Retrieve City Boundaries
    city_boundaries = ox.geocode_to_gdf(f"{city_name}, Germany")

    # Retrieve the shape of the specified postal code within the city
    postal_code_shape = ox.features_from_place(f"{city_name}, Germany", tags={"boundary": "postal_code", "postal_code": postal_code})

    # Plot the city boundaries along with the shape of the specified postal code
    ax = city_boundaries.plot(color='lightgrey', edgecolor='black')
    postal_code_shape.plot(ax=ax, color='none', edgecolor='red', linewidth=2)
    plt.title(f"City Boundaries with Postal Code - {postal_code}, {city_name}, Germany")
    plt.show()

# Example usage
city_name = "Kassel"
postal_code = "34127"
plot_postal_code_shape(city_name, postal_code)
```

```python
import osmnx as ox
import geopandas as gpd
import matplotlib.pyplot as plt

def get_zip_code_borders(postal_codes):
    # Initialize an empty GeoDataFrame
    zip_code_gdf = gpd.GeoDataFrame()

    for postal_code in postal_codes:
        # Retrieve the shape of the specified postal code(s) using address
        address = f"{postal_code}, Germany"
        zip_code_shape = ox.features_from_address(address, tags={"boundary": "postal_code", "postal_code": postal_code})

        # Concatenate the retrieved shape to the GeoDataFrame
        zip_code_gdf = pd.concat([zip_code_gdf, zip_code_shape])

    return zip_code_gdf.unary_union

def plot_zip_code_borders(postal_codes):
    # Retrieve the shape of the specified postal code(s)
    zip_code_borders = get_zip_code_borders(postal_codes)
    
    # Plot the shape of the specified postal code(s)
    ax = gpd.GeoSeries(zip_code_borders).plot(color='none', edgecolor='red', linewidth=2)

    plt.title(f"Postal Code Boundaries - {postal_codes}, Germany")
    plt.show()

# Example usage
postal_codes_to_plot = ["34127"]

# Get and print the shape of the postal code(s)
zip_code_borders = get_zip_code_borders(postal_codes_to_plot)
print(zip_code_borders)

# Plot the shape of the postal code(s)
plot_zip_code_borders(postal_codes_to_plot)
```
